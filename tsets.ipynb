```python
import pandas as pd
import altair as alt
alt.data_transformers.enable('data_server')
```

# Data

```python
# Proposals, in chronological order.
props = pd.read_csv('props.csv')
props = props[['ledger', 'seq', 'node', 'tset']]
props
```

```python
# Ledgers, in chronological order.
ledgers = props['ledger'].drop_duplicates(keep='first').reset_index(drop=True)
ledgers
```

```python
# Buckets for histograms.
n = 10
buckets = ledgers.index // (ledgers.size // n)
buckets = pd.DataFrame({'ledger': ledgers, 'bucket': buckets})
buckets
```

```python
# Transaction sets.
tsets = pd.read_csv('tsets.csv')
tsets = tsets.drop_duplicates()
tsets
```

```python
# Transaction set sizes.
sizes = tsets.groupby('tset', sort=False)['txn'].count().sort_values()
sizes.describe()
```

```python
# The validated transaction set per ledger (i.e. the most proposed transaction set).
vtset = props.groupby(['ledger', 'tset'], sort=False)['node'].count().reset_index()
vtset = vtset[vtset['node'] > 34 * 0.6]
vtset = vtset[vtset['node'] == vtset.groupby(['ledger'], sort=False)['node'].transform(max)]
vtset
```

```python
# The validated transactions per ledger.
vtxns = vtset.drop(columns=['node']).merge(tsets, on='tset').drop(columns=['tset'])
vtxns
```

```python
# Proposals missing transaction sets.
proposal_incomplete = ~props['tset'].isin(tsets['tset'].unique())
proposal_incomplete.sum()
```

```python
# Number of proposals per election that are missing transaction sets.
incomplete_elections = props[proposal_incomplete]['ledger'].value_counts()
incomplete_elections.describe()
```

```python
# Complete elections.
complete_elections = ledgers[~ledgers.isin(incomplete_elections.index)]
complete_elections
```

```python
# The proposed transactions per ledger.
ptxns = props[props['seq'] == 0].drop(columns='seq')
ptxns = ptxns.merge(tsets, on='tset').drop(columns=['tset'])
ptxns
```

```python
# Initial support for each transaction per ledger.
support = ptxns.groupby(['ledger', 'txn'], sort=False)['node'].count().reset_index()
support
```

# Are validators including enough transactions in first proposals?

```python
# Distribution of proposal sizes per round.
xs = props.merge(sizes, on='tset')
xs.groupby('seq')['txn'].describe()
```

```python
# Distribution of sizes of first proposals per validator.
xs = props[props['seq'] == 0].merge(sizes, on='tset').drop(columns='seq')
xs.groupby(['node'])['txn'].describe()
```

```python
# The nodes proposing the fewest transactions per ledger.
nodes = ptxns.groupby(['ledger', 'node'], sort=False)['txn'].count()
nodes = nodes.groupby('ledger', group_keys=False, sort=False).nsmallest(1)
nodes.reset_index()['node'].value_counts()
```

```python
# Distribution of sizes of first proposal over time per validator.
data = props[props['seq'] == 0].drop(columns='seq')
data = data.merge(buckets, on='ledger').merge(sizes, on='tset')
data = data.groupby(['node', 'bucket']).describe()
data.columns = data.columns.droplevel()
data = data.reset_index()
data['node'] = data['node'].str.slice(stop=10)

xaxis = alt.Chart(data).encode(
    x=alt.X('bucket:O', axis=alt.Axis(title='ledger', labels=False, ticks=False)),
)

tooltip = ['max', '75%', '50%', '25%', 'min']

wick = xaxis.mark_rule().encode(
    alt.Y('min:Q', title='transactions'),
    alt.Y2('max:Q'),
    tooltip=tooltip,
)

candle = xaxis.mark_bar(fill='white', stroke='#4c78a8').encode(
    alt.Y('25%:Q'),
    alt.Y2('75%:Q'),
    tooltip=tooltip,
)

(
    wick + candle
).properties(
    width=160,
    height=160,
).facet(
    'node',
    columns=5,
)
```

# Are validators including low-fee transactions in first proposals?

```python
# Number of low-fee transactions per first proposal over time per validator.
data = props[props['seq'] == 0].drop(columns='seq')
data = data.merge(buckets, on='ledger')
ns = tsets[tsets['fee'] < 20].groupby('tset')['txn'].count()
data = data.merge(ns, on='tset')
data = data.groupby(['node', 'bucket'])['txn'].describe()
data = data.reset_index()
data['node'] = data['node'].str.slice(stop=10)

xaxis = alt.Chart(data).encode(
    x=alt.X('bucket:O', axis=alt.Axis(title='ledger', labels=False, ticks=False)),
)

tooltip = ['max', '75%', '50%', '25%', 'min']

wick = xaxis.mark_rule().encode(
    alt.Y('min:Q', title='low-fee transactions'),
    alt.Y2('max:Q'),
    tooltip=tooltip,
)

candle = xaxis.mark_bar(fill='white', stroke='#4c78a8').encode(
    alt.Y('25%:Q'),
    alt.Y2('75%:Q'),
    tooltip=tooltip,
)

(
    wick + candle
).properties(
    width=160,
    height=160,
).facet(
    'node',
    columns=5,
)
```

# Are validators re-proposing deferred transactions?

```python
# Deferred transactions per ledger.
dtxns = ptxns[['ledger', 'txn']].drop_duplicates()
dtxns = dtxns.merge(vtxns.assign(validated=True).drop(columns='fee'), on=['ledger', 'txn'], how='left')
dtxns['validated'] = dtxns['validated'].fillna(False)
dtxns = dtxns[~dtxns['validated']].drop(columns='validated')
dtxns
```

```python
# Distribution of number of deferred transactions per ledger.
xs = dtxns.groupby('ledger').count()
xs['txn'] = xs['txn'] // 100 * 100
xs = xs.reset_index().groupby('txn')['ledger'].count().reset_index()
alt.Chart(xs).mark_bar().encode(
    x=alt.X('txn', bin=alt.Bin(step=100)),
    y='ledger',
)
```

```python
parents = pd.DataFrame({'child': ledgers, 'parent': pd.Series(ledgers.values, index=ledgers.index + 1)})
parents
```

```python
# What is the support for each transaction in the ledger after it was deferred?
rptxns = dtxns.merge(parents.dropna(), left_on='ledger', right_on='parent').drop(columns=['ledger'])
rptxns = rptxns.merge(support, left_on=['child', 'txn'], right_on=['ledger', 'txn'], how='left')
rptxns['node'] = rptxns['node'].fillna(0, downcast='infer')
rptxns
```

```python
# Distribution of support for deferred transactions in subsequent proposals.
xs = rptxns.groupby('node')['txn'].count().reset_index()
alt.Chart(xs).mark_bar().encode(
    x=alt.X('node', bin=alt.Bin(step=1)),
    y='txn',
)
```
