```python
import pandas as pd
import altair as alt
alt.data_transformers.enable('data_server')
```

# Data

```python
# Proposals, in chronological order.
props = pd.read_csv('props.csv')
props = props[['ledger', 'seq', 'node', 'tset']]
props
```

```python
# Ledgers, in chronological order.
ledgers = props['ledger'].drop_duplicates(keep='first').reset_index(drop=True)
ledgers
```

```python
# Transaction sets.
tsets = pd.read_csv('tsets.csv')
tsets = tsets.drop_duplicates()
tsets
```

```python
# Transaction set sizes.
sizes = tsets.groupby('tset', sort=False)['txn'].count().sort_values()
sizes.describe()
```

```python
# The validated transaction set per ledger (i.e. the most proposed transaction set).
vtset = props.groupby(['ledger', 'tset'], sort=False)['node'].count().reset_index()
vtset = vtset[vtset['node'] > 34 * 0.6]
vtset = vtset[vtset['node'] == vtset.groupby(['ledger'], sort=False)['node'].transform(max)]
vtset
```

```python
# The validated transactions per ledger.
vtxns = vtset.drop(columns=['node']).merge(tsets, on='tset').drop(columns=['tset'])
vtxns
```

```python
# Proposals missing transaction sets.
proposal_incomplete = ~props['tset'].isin(tsets['tset'].unique())
proposal_incomplete.sum()
```

```python
# Number of proposals per election that are missing transaction sets.
incomplete_elections = props[proposal_incomplete]['ledger'].value_counts()
incomplete_elections.describe()
```

```python
# Complete elections.
complete_elections = ledgers[~ledgers.isin(incomplete_elections.index)]
complete_elections
```

```python
# The proposed transactions per ledger.
ptxns = props[(props['seq'] == 0) & props['ledger'].isin(complete_elections)].drop(columns='seq')
ptxns = ptxns.merge(tsets, on='tset', sort=False).drop(columns=['tset'])
ptxns
```

# Are validators including enough transactions in first proposals?

```python
# Distribution of proposal sizes per round.
xs = props.merge(sizes, on='tset', sort=False)
xs.groupby('seq')['txn'].describe()
```

```python
# Distribution of sizes of first proposals per validator.
xs = props[props['seq'] == 0].merge(sizes, on='tset', sort=False).drop(columns='seq')
xs.groupby(['node'])['txn'].describe()
```

```python
# Distribution of sizes of first proposal over time per validator.

nbuckets = 10
bucket_size = ledgers.size // nbuckets
buckets = ledgers.index // bucket_size
bs = pd.DataFrame({'ledger': ledgers, 'bucket': buckets})

def quantile(n):
    def quantile(df):
        return df.quantile(n)
    quantile.__name__ = 'quantile_{:2.0f}'.format(n*100)
    return quantile

xs = props[props['seq'] == 0].drop(columns='seq')
xs = xs.merge(bs, on='ledger').merge(sizes, on='tset')
xs = xs.groupby(['node', 'bucket']).describe()
xs.columns = xs.columns.droplevel()
xs = xs.reset_index()
xs['node'] = xs['node'].str.slice(stop=10)


xaxis = alt.Chart(xs).encode(
    x=alt.X('bucket:O', axis=alt.Axis(title='ledger', labels=False, ticks=False)),
)

tooltip = ['max', '75%', '50%', '25%', 'min']

wick = xaxis.mark_rule().encode(
    alt.Y('min:Q', title='transactions'),
    alt.Y2('max:Q'),
    tooltip=tooltip,
)

candle = xaxis.mark_bar(fill='white', stroke='#4c78a8').encode(
    alt.Y('25%:Q'),
    alt.Y2('75%:Q'),
    tooltip=tooltip,
)

(
    wick + candle
).properties(
    width=160,
    height=160,
).facet(
    'node',
    columns=5,
)
```

# Are validators re-proposing deferred transactions?

```python
# The deferred transactions per ledger.
dtxns = ptxns[~ptxns.set_index(['ledger', 'txn']).index.isin(vtxns.set_index(['ledger', 'txn']).index)]
dtxns
```

```python
# Number of transaction sets recorded.
tsets['tset'].nunique()
```

```python
# The deferred transactions with more than one proposer.
ts = dtxns.groupby(['ledger', 'txn'])['node'].count()
ts = ts[ts > 1]
ts = ts.groupby(['ledger'], group_keys=False).nlargest(3)
ts = ts.reset_index().sort_values('node')
ts
```

```python
# The number of transactions proposed and deferred per proposer per election.
txns = ptxns.groupby(['ledger', 'node'])['txn'].count().rename('proposed').reset_index()
txns = txns.merge(vtset[['ledger', 'txn']], on='ledger').rename(columns={'txn': 'validated'})
txns['deferred'] = txns['proposed'] - txns['validated']
txns
```

```python
# The range of proposal sizes.
txns.sort_values('proposed').iloc[[0, -1]]
```

```python
# The nodes with the fewest deferred transactions per ledger.
winners = txns.loc[txns.groupby('ledger')['deferred'].idxmin()]
winners
```

```python
# Are these two sets the same?
(nodes.set_index(['ledger', 'node']).index == winners.set_index(['ledger', 'node']).index).all()
```

```python
# Who are the culprits?
nodes.groupby('node')['proposed'].count().sort_values()
```

```python
# What are the minimum fees in these proposals?
smallest = nodes.merge(props[props['seq'] == 0], on=['ledger', 'node']).drop(columns=['seq'])
smallest = smallest.merge(tsets.groupby('tset')['fee'].min(), on='tset')
smallest
```

```python
# What is the minimum fee among smallest first proposals per node?
smallest.groupby('node')['fee'].min()
```
